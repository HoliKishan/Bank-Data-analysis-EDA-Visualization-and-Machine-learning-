# ***Train Test Split***

```python
from sklearn.model_selection import train_test_split # Import train and split function from sklearn lib.
```

```python
X_train, X_test, y_train, y_test =train_test_split(x,y,test_size=0.3) # Spliting 70% data for training and 30% data for testing and store values in variable
```

```python
print(X_train.shape)  #Print the shape of training dataframe
```

```python
print(X_test.shape) #Print the shape of testing dataframe
```

# ***Logistic Regression***

```python
from sklearn.linear_model import LogisticRegression  # Import Logistic Regression(Machine-learning)  from Skitlearn library
```

```python
logmodel = LogisticRegression()  # Create a extension of model
```

```python
logmodel.fit(X_train,y_train)   # fit the model on x-train and ytrain
```

```python
predictions = logmodel.predict(X_test)   #Predict the value on test dataset
```

```python
from sklearn.metrics import classification_report  # model Evalation
```

```python
print(classification_report(y_test,predictions)) # Check accuracy of model
```

```python
from sklearn.metrics import confusion_matrix # import confusion matrix from skitlear library
```

```python
print(confusion_matrix(y_test,predictions))  #View confusion matrix
```
